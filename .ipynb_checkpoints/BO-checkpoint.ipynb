```python
# Force Field: MMFF94
ff = pybel._forcefields['mmff94']

def energy():
    return ff.Energy(False)
```

```python
def torsion_potential(x, para):
    V = para[1]*(1 + para[0]*math.cos(x)) + para[3]*(1 + para[2]*math.cos(2*x)) + para[5]*(1 + para[4]*math.cos(x))  
    + para[7]*(1 + para[6]*math.cos(x)) + para[9]*(1 + para[8]*math.cos(x)) + para[11]*(1 + para[10]*math.cos(x))
    return(V)
```

```python
x = np.linspace(0, 2*math.pi)
potential_1 = map(lambda i: torsion_potential(i, np.array(torsion_pref.iloc[141,1:])), x)
potential_2 = map(lambda i: torsion_potential(i, np.array(torsion_pref.iloc[151,1:])), x )
```

```python
# Function to calculate energy 
# Input torsion angles
def calculate_energy(torsion_angles):
    ff = pybel._forcefields['mmff94']
    output = []
    for j in range(torsion_angles.shape[0]):
        for i in range(torsion_angles.shape[1]):
            rotors[i].SetToAngle(currentCoords, torsion_angles[j,i])
            molecule.OBMol.SetCoordinates(currentCoords)
            ff.SetCoordinates(molecule.OBMol)
        output.append(ff.Energy())
    return np.array(output).reshape([torsion_angles.shape[0],1])
```

```python
seed(123)

# Set Kernel  
Periodic = GPy.kern.StdPeriodic(input_dim=2,period=[2,2],lengthscale=[1,1], ARD1=True, ARD2=True)
rbf = GPy.kern.RBF(input_dim = 2, ARD=True, lengthscale=[1,1])
locally_periodic = GPy.kern.Prod([Periodic, rbf])

# Setup GP Model
model = GPyOpt.models.GPModel(kernel=locally_periodic,exact_feval=True,optimize_restarts=10,verbose=False)

# Objective Function
objective = GPyOpt.core.task.SingleObjective(calculate_energy)

# Space design
temp = [{'name': 'torsion', 'type': 'continuous', 'domain': (0, 2*math.pi)}]
domain = temp*rl.Size()
space = GPyOpt.Design_space(domain)

# Set Aquisition Optimizer 
aquisition_optimizer = GPyOpt.optimization.AcquisitionOptimizer(space)

# Give initial samples by random sampling (5 samples) in the space
initial_design = GPyOpt.experiment_design.initial_design('random', space, 5)

# Use EI as acqusition function
acquisition = GPyOpt.acquisitions.AcquisitionEI(model, space, optimizer=aquisition_optimizer)

# Sequential Evaulation
evaluator = GPyOpt.core.evaluators.Sequential(acquisition)

# Optimization Step
bo = GPyOpt.methods.ModularBayesianOptimization(model, space, objective, acquisition, evaluator, initial_design)

# Budget
max_iter  = 50

bo.run_optimization(max_iter = max_iter)

bo.plot_convergence()
print(bo.fx_opt, bo.x_opt)
```

```python
# Compare with the 'best '
mmff_pes = pd.read_table('twist-mmff-pes.csv', delimiter=',')
```

```python
# Search for the lowest energy
location = np.argmin(mmff_pes.iloc[:,2])
best_angles = map(lambda x: math.radians(x), mmff_pes.iloc[location,:2])
```

```python
best_mol = molecule.OBMol
simulated_mol = molecule.OBMol

Coords_1 = molecule.OBMol.GetCoordinates()
Coords_2 = molecule.OBMol.GetCoordinates()
align = pybel.ob.OBAlign()
for i in range(len(best_angles)):
    rotors[i].SetToAngle(Coords_1, best_angles[i])
    best_mol.SetCoordinates(Coords_1)

align.SetRefMol(best_mol)

for i in range(len(bo.x_opt)):
    rotors[i].SetToAngle(Coords_2, bo.x_opt[i])
    simulated_mol.SetCoordinates(Coords_2)

align.SetTargetMol(simulated_mol)
align.Align()
align.GetRMSD()
```
